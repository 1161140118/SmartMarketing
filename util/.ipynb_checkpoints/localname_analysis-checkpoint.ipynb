```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

```python
localname = pd.read_csv("../data/raw_data/localname.csv",names=['store','labels','addr','lng','lat','lnum']).drop_duplicates()
# 去重前 19w， 去重后12w
```

```python
localname.groupby( by=localname.labels.apply(lambda x : x.split(';')[0]) ).count()
```

```python
# 过滤无效地址 并提取 '购物服务'
# addr = '[]' 2772个
store_name = localname[ (localname.addr != '[]') & (localname.labels.apply(lambda x : x.split(';')[0] in ['购物服务'])) ]
```

```python
store_name['label2'] = store_name.labels.apply(lambda x : x.split(';')[1])
store_name['label3'] = store_name.labels.apply(lambda x : x.split('|')[0].split(';')[2])
```

```python
store_name.groupby(['label2','label3']).count()
store_name.groupby(['label2','label3']).count().to_csv("../data/shop_cnt.csv")
```

```python
store_name.groupby(
    by= [ store_name.labels.apply(lambda x : x.split(';')[1]),store_name.labels.apply(lambda x : x.split(';')[2])]
    ).count().sort_values(by='store',ascending=False)
```

```python
def label_filter(x):
    if x['store'][-2:] is '商行':
        return False
    if x['label2'] in ['商场','个人用品/化妆品店','服装鞋帽皮具店','体育用品店']:
        return True
    if x['label2'] is '特色商业街' and x['label3'] in ['步行街','特色商业街']:
        return True
    if x['label2'] is '专卖店' and x['lable3'] in ['儿童用品店','礼品饰品店','珠宝首饰工艺品']:
        return True
    return False
```

```python
store_name = store_name[ store_name.apply(label_filter,axis=1) ]
```

```python
store_name.groupby(["label2"]).count()
```

```python
store_name.groupby(["label2","label3"]).count()
```

```python
store_name.to_csv("../data/shop.csv", index=None)
```
