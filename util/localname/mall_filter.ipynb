```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #解决plt中文乱码问题
```

```python
mall = pd.read_csv("data/mall.csv")
```

```python
def area_define(x):
    """
    1. 江北松北城区：lng[126.457, 126.565],lat[45.795, 45.837]
    2. 江北师大城区：lng[126.550, 126.605],lat[45.860, 45.895]  // 126.549735,45.860136  126.604838,45.894853
    3. 江南中心城区：lng[126.522, 126.707],lat[45.691, 45.801]  // 126.522612,45.691439 126.706976,45.800928
    """
    if 126.457 < x.lng < 126.565 and 45.795 < x.lat < 45.837 :
        return "松北"
    if 126.550 < x.lng < 126.605 and 45.680 < x.lat < 45.895 :
        return "师大"
    if 126.522 < x.lng < 126.707 and 45.691 < x.lat < 45.801 :
        return "中心"
    return "其他"
mall['area'] = mall.apply(lambda x : area_define(x), axis = 1)
```
